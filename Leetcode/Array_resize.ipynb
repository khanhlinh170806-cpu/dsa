```python
# Remove Element

def removeElement(nums, val):
    k = 0  # vị trí ghi phần tử hợp lệ

    for i in range(len(nums)):
        if nums[i] != val:
            nums[k] = nums[i]  # overwrite
            k += 1

    return k


# Test
nums = [3, 2, 2, 3, 4, 3]
val = 3

new_length = removeElement(nums, val)

print("New length:", new_length)
print("Array after operation:", nums[:new_length])
```

```python
# Merge Sorted Array

def merge(nums1, m, nums2, n):
    i = m - 1
    j = n - 1
    k = m + n - 1

    while i >= 0 and j >= 0:
        if nums1[i] > nums2[j]:
            nums1[k] = nums1[i]
            i -= 1
        else:
            nums1[k] = nums2[j]
            j -= 1
        k -= 1

    while j >= 0:
        nums1[k] = nums2[j]
        j -= 1
        k -= 1


# Test
nums1 = [1, 2, 3, 0, 0, 0]
m = 3
nums2 = [2, 5, 6]
n = 3

merge(nums1, m, nums2, n)

print("Merged array:", nums1)
```

```python
# Move Zeroes

def moveZeroes(nums):
    k = 0  # vị trí ghi số khác 0

    # đưa các phần tử khác 0 lên đầu
    for i in range(len(nums)):
        if nums[i] != 0:
            nums[k] = nums[i]
            k += 1

    # điền 0 vào phần còn lại
    for i in range(k, len(nums)):
        nums[i] = 0


# Test
nums = [0, 1, 0, 3, 12]

moveZeroes(nums)

print("After moving zeroes:", nums)
```
