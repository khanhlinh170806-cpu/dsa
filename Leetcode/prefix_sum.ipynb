1. RUNNING SUM OF 1D ARRAY

```python
# ============================================================
# LeetCode 1480 - Running Sum of 1D Array
# Bài toán: Cho mảng nums, trả về mảng running sum.
# running_sum[i] = nums[0] + nums[1] + ... + nums[i]
# ============================================================

from typing import List


def running_sum(nums: List[int]) -> List[int]:
    """
    Tính running sum (tổng tích lũy) của mảng 1 chiều.

    Args:
        nums: Mảng số nguyên đầu vào.

    Returns:
        Mảng running sum cùng độ dài với nums.
    """

    # Duyệt qua từng vị trí từ index 1 đến cuối mảng
    # (index 0 giữ nguyên vì chưa có phần tử nào trước nó)
    for i in range(1, len(nums)):

        # Cộng dồn: phần tử hiện tại += phần tử liền trước
        # Sau bước này, nums[i] chứa tổng từ nums[0] đến nums[i]
        nums[i] += nums[i - 1]

    # Trả về mảng đã được biến đổi thành running sum
    return nums


# ============================================================
# Test cases
# ============================================================

def run_tests():
    test_cases = [
        {
            "input": [1, 2, 3, 4],
            "expected": [1, 3, 6, 10],
            "desc": "Mảng tăng dần thông thường"
        },
        {
            "input": [1, 1, 1, 1, 1],
            "expected": [1, 2, 3, 4, 5],
            "desc": "Tất cả phần tử bằng 1"
        },
        {
            "input": [3, 1, 2, 10, 1],
            "expected": [3, 4, 6, 16, 17],
            "desc": "Mảng hỗn hợp"
        },
        {
            "input": [5],
            "expected": [5],
            "desc": "Mảng chỉ có 1 phần tử"
        },
        {
            "input": [0, 0, 0],
            "expected": [0, 0, 0],
            "desc": "Toàn số 0"
        },
    ]

    print("=" * 50)
    print("  Chạy test: Running Sum of 1D Array")
    print("=" * 50)

    passed = 0
    for i, tc in enumerate(test_cases):
        # Tạo bản sao để không làm thay đổi dữ liệu gốc khi so sánh
        input_copy = tc["input"][:]
        result = running_sum(input_copy)
        ok = result == tc["expected"]

        status = "PASS" if ok else "FAIL"
        if ok:
            passed += 1

        print(f"\nTest {i + 1}: {tc['desc']}")
        print(f"  Input    : {tc['input']}")
        print(f"  Expected : {tc['expected']}")
        print(f"  Got      : {result}")
        print(f"  Result   : [{status}]")

    print("\n" + "=" * 50)
    print(f"  Tổng kết: {passed}/{len(test_cases)} test passed")
    print("=" * 50)


if __name__ == "__main__":
    run_tests()
```

2. FIND PIVOT INDEX

```python
# ============================================================
# LeetCode 724 - Find Pivot Index
# Bài toán: Tìm pivot index i sao cho:
#   sum(nums[0..i-1]) == sum(nums[i+1..n-1])
# Nếu không tìm được, trả về -1.
# ============================================================

from typing import List


def pivot_index(nums: List[int]) -> int:
    """
    Tìm chỉ số pivot của mảng.

    Ý tưởng:
        - Tính tổng toàn bộ mảng trước (total).
        - Duyệt từng index i, duy trì biến left_sum tích lũy.
        - Tổng bên phải = total - left_sum - nums[i]
        - Nếu left_sum == right_sum thì i là pivot.

    Độ phức tạp: O(n) thời gian, O(1) bộ nhớ.

    Args:
        nums: Mảng số nguyên đầu vào.

    Returns:
        Index pivot đầu tiên tìm được, hoặc -1 nếu không có.
    """

    # Tính tổng toàn bộ mảng một lần
    total = sum(nums)

    # Biến lưu tổng các phần tử bên trái index hiện tại
    # Ban đầu chưa có phần tử nào bên trái → left_sum = 0
    left_sum = 0

    for i in range(len(nums)):

        # Tổng bên phải = tổng toàn bộ - phần tử hiện tại - tổng bên trái
        right_sum = total - nums[i] - left_sum

        # Kiểm tra điều kiện pivot
        if left_sum == right_sum:
            return i  # Trả về ngay index đầu tiên tìm được

        # Cộng nums[i] vào left_sum để chuẩn bị cho vòng lặp tiếp theo
        left_sum += nums[i]

    # Duyệt hết mảng mà không tìm được pivot → trả về -1
    return -1


# ============================================================
# Test cases
# ============================================================

def run_tests():
    test_cases = [
        {
            "input": [1, 7, 3, 6, 5, 6],
            "expected": 3,
            "desc": "Pivot ở giữa mảng (index 3)"
        },
        {
            "input": [1, 2, 3],
            "expected": -1,
            "desc": "Không có pivot"
        },
        {
            "input": [2, 1, -1],
            "expected": 0,
            "desc": "Pivot ở đầu mảng (index 0, bên trái rỗng = 0)"
        },
        {
            "input": [-1, -1, -1, 0, 1, 1],
            "expected": 0,
            "desc": "Mảng có số âm, pivot ở index 0"
        },
        {
            "input": [0],
            "expected": 0,
            "desc": "Mảng 1 phần tử — luôn là pivot"
        },
        {
            "input": [0, 0, 0, 0],
            "expected": 0,
            "desc": "Toàn số 0 — pivot đầu tiên là index 0"
        },
        {
            "input": [-1, -1, 0, 1, 1, 0],
            "expected": 5,
            "desc": "Pivot ở cuối mảng (bên phải rỗng = 0)"
        },
    ]

    print("=" * 55)
    print("  Chạy test: Find Pivot Index")
    print("=" * 55)

    passed = 0
    for i, tc in enumerate(test_cases):
        result = pivot_index(tc["input"][:])  # bản sao để giữ input gốc
        ok = result == tc["expected"]
        if ok:
            passed += 1

        status = "PASS" if ok else "FAIL"
        print(f"\nTest {i + 1}: {tc['desc']}")
        print(f"  Input    : {tc['input']}")
        print(f"  Expected : {tc['expected']}")
        print(f"  Got      : {result}")
        print(f"  Result   : [{status}]")

    print("\n" + "=" * 55)
    print(f"  Tổng kết: {passed}/{len(test_cases)} test passed")
    print("=" * 55)


if __name__ == "__main__":
    run_tests()
```

3. RANGE SUM QUERY - IMMUTABLE

```python
# ============================================================
# LeetCode 303 - Range Sum Query - Immutable
# Bài toán: Cho mảng nums (không thay đổi), xử lý nhiều
#   truy vấn sumRange(left, right) trả về:
#   nums[left] + nums[left+1] + ... + nums[right]
#
# Ý tưởng: Xây dựng mảng prefix_sum một lần trong __init__,
#   sau đó mỗi truy vấn chỉ cần O(1) bằng phép trừ đơn giản.
#
# prefix_sum[i] = nums[0] + nums[1] + ... + nums[i-1]
#   (shift 1 vị trí để tránh check biên khi i = 0)
#
# sumRange(l, r) = prefix_sum[r+1] - prefix_sum[l]
# ============================================================

from typing import List


class NumArray:

    def __init__(self, nums: List[int]):
        """
        Khởi tạo cấu trúc dữ liệu với mảng prefix sum.

        prefix_sum có độ dài len(nums) + 1.
        prefix_sum[0] = 0  (phần tử canh gác, không có phần tử nào trước index 0)
        prefix_sum[i] = tổng của nums[0..i-1]
        """

        n = len(nums)

        # Tạo mảng prefix_sum với n+1 phần tử, khởi tạo toàn 0
        self.prefix_sum = [0] * (n + 1)

        # Điền prefix_sum: mỗi phần tử = phần tử trước + nums[i]
        for i in range(n):
            self.prefix_sum[i + 1] = self.prefix_sum[i] + nums[i]

        # Ví dụ với nums = [1, 7, 3, 6]:
        # prefix_sum = [0, 1, 8, 11, 17]

    def sumRange(self, left: int, right: int) -> int:
        """
        Tính tổng nums[left..right] trong O(1).

        Công thức:
            sum(left, right) = prefix_sum[right+1] - prefix_sum[left]

        Tại sao đúng?
            prefix_sum[right+1] = nums[0] + ... + nums[right]
            prefix_sum[left]     = nums[0] + ... + nums[left-1]
            Hiệu số               = nums[left] + ... + nums[right]  ✓

        Args:
            left:  Index bắt đầu (inclusive).
            right: Index kết thúc (inclusive).

        Returns:
            Tổng các phần tử từ left đến right.
        """
        return self.prefix_sum[right + 1] - self.prefix_sum[left]


# ============================================================
# Test cases
# ============================================================

def run_tests():
    test_cases = [
        {
            "nums": [-2, 0, 3, -5, 2, -1],
            "queries": [
                ((0, 2),  1,  "Tổng 3 phần tử đầu: -2+0+3"),
                ((2, 5), -1,  "Tổng từ index 2 đến 5: 3-5+2-1"),
                ((0, 5), -3,  "Tổng toàn mảng"),
            ],
        },
        {
            "nums": [1, 7, 3, 6, 5, 6],
            "queries": [
                ((0, 0),  1,  "Chỉ 1 phần tử đầu"),
                ((1, 3), 16,  "7+3+6"),
                ((0, 5), 28,  "Tổng toàn mảng"),
            ],
        },
        {
            "nums": [0],
            "queries": [
                ((0, 0),  0,  "Mảng 1 phần tử bằng 0"),
            ],
        },
        {
            "nums": [-5, -3, -1],
            "queries": [
                ((0, 2), -9, "Toàn số âm, tổng toàn mảng"),
                ((1, 2), -4, "2 phần tử cuối"),
            ],
        },
    ]

    print("=" * 58)
    print("  Chạy test: Range Sum Query - Immutable")
    print("=" * 58)

    total_pass = 0
    total_count = 0

    for group in test_cases:
        nums = group["nums"]
        obj = NumArray(nums)  # Khởi tạo một lần cho cả nhóm query

        print(f"\nnums = {nums}")
        print(f"prefix_sum = {obj.prefix_sum}")
        print("-" * 45)

        for (left, right), expected, desc in group["queries"]:
            result = obj.sumRange(left, right)
            ok = result == expected
            if ok:
                total_pass += 1
            total_count += 1

            status = "PASS" if ok else "FAIL"
            print(f"  sumRange({left}, {right})")
            print(f"    {desc}")
            print(f"    Expected: {expected}  |  Got: {result}  [{status}]")

    print("\n" + "=" * 58)
    print(f"  Tổng kết: {total_pass}/{total_count} test passed")
    print("=" * 58)


if __name__ == "__main__":
    run_tests()
```

4. SUBARRAY SUM EQUALS K

```python
# ============================================================
# LeetCode 560 - Subarray Sum Equals K
# Bài toán: Đếm số subarray liên tiếp có tổng bằng k.
#
# Ý tưởng: Dùng prefix sum + hash map (dictionary).
#
# Nếu prefix_sum[j] - prefix_sum[i] == k
# thì subarray nums[i..j-1] có tổng bằng k.
# → Tức là: prefix_sum[i] == prefix_sum[j] - k
#
# Thay vì lưu toàn bộ mảng prefix_sum, ta chỉ cần đếm
# tần suất từng giá trị prefix_sum đã gặp bằng dict.
#
# Độ phức tạp: O(n) thời gian, O(n) bộ nhớ.
# ============================================================

from typing import List
from collections import defaultdict


def subarray_sum(nums: List[int], k: int) -> int:
    """
    Đếm số subarray có tổng bằng k.

    Args:
        nums: Mảng số nguyên (có thể âm).
        k:    Giá trị tổng cần tìm.

    Returns:
        Số lượng subarray có tổng bằng k.
    """

    # dict đếm số lần mỗi giá trị prefix_sum đã xuất hiện.
    # Khởi tạo {0: 1} vì prefix_sum = 0 xuất hiện 1 lần
    # trước khi duyệt (tương đương subarray bắt đầu từ index 0).
    prefix_count = defaultdict(int)
    prefix_count[0] = 1

    current_sum = 0  # prefix sum tích lũy đến vị trí hiện tại
    count = 0        # số subarray thỏa điều kiện

    for num in nums:

        # Cộng dồn phần tử hiện tại vào prefix sum
        current_sum += num

        # Nếu (current_sum - k) đã từng xuất hiện trước đó,
        # nghĩa là tồn tại một điểm i sao cho:
        #   prefix_sum[i] == current_sum - k
        #   → subarray từ i đến vị trí hiện tại có tổng = k
        # Số lượng điểm i như vậy = prefix_count[current_sum - k]
        count += prefix_count[current_sum - k]

        # Ghi nhận prefix sum hiện tại vào dict
        # (thực hiện SAU khi cộng vào count để tránh dùng chính mình)
        prefix_count[current_sum] += 1

    return count


# ============================================================
# Test cases
# ============================================================

def run_tests():
    test_cases = [
        {
            "nums": [1, 1, 1],
            "k": 2,
            "expected": 2,
            "desc": "Hai subarray [1,1] có tổng = 2: index (0,1) và (1,2)"
        },
        {
            "nums": [1, 2, 3],
            "k": 3,
            "expected": 2,
            "desc": "[3] và [1,2] đều có tổng = 3"
        },
        {
            "nums": [1, -1, 1],
            "k": 1,
            "expected": 3,
            "desc": "Mảng có số âm: [1], [1,-1,1], [1]"
        },
        {
            "nums": [1, 2, 3, -3, 3],
            "k": 3,
            "expected": 5,
            "desc": "Nhiều subarray với số âm bù trừ: [1,2],[1,2,3,-3],[3],[3,-3,3],[3]"
        },
        {
            "nums": [0, 0, 0],
            "k": 0,
            "expected": 6,
            "desc": "Toàn số 0, k=0: mọi subarray đều thỏa"
        },
        {
            "nums": [3],
            "k": 3,
            "expected": 1,
            "desc": "Mảng 1 phần tử bằng k"
        },
        {
            "nums": [3],
            "k": 5,
            "expected": 0,
            "desc": "Mảng 1 phần tử, không có subarray nào thỏa"
        },
        {
            "nums": [-1, -1, 1],
            "k": -1,
            "expected": 3,
            "desc": "k âm: [-1], [-1], [-1,-1,1] đều có tổng = -1"
        },
    ]

    print("=" * 58)
    print("  Chạy test: Subarray Sum Equals K")
    print("=" * 58)

    passed = 0
    for i, tc in enumerate(test_cases):
        result = subarray_sum(tc["nums"][:], tc["k"])
        ok = result == tc["expected"]
        if ok:
            passed += 1

        status = "PASS" if ok else "FAIL"
        print(f"\nTest {i + 1}: {tc['desc']}")
        print(f"  nums     : {tc['nums']}")
        print(f"  k        : {tc['k']}")
        print(f"  Expected : {tc['expected']}")
        print(f"  Got      : {result}")
        print(f"  Result   : [{status}]")

    print("\n" + "=" * 58)
    print(f"  Tổng kết: {passed}/{len(test_cases)} test passed")
    print("=" * 58)


if __name__ == "__main__":
    run_tests()
```

5. CONTIGUOUS ARRAY

```python
# ============================================================
# LeetCode 525 - Contiguous Array
# Bài toán: Tìm độ dài lớn nhất của subarray có số lượng 0 và 1 bằng nhau.
#
# Ý tưởng: Biến bài toán về prefix sum.
#
# Ta coi:
#   0 -> -1
#   1 -> +1
#
# Khi đó:
# Nếu prefix_sum[j] == prefix_sum[i]
# → tổng từ i+1 đến j = 0
# → số lượng 0 và 1 bằng nhau
#
# Ta lưu vị trí đầu tiên xuất hiện của mỗi prefix_sum.
# Khi gặp lại cùng giá trị → cập nhật độ dài lớn nhất.
#
# Độ phức tạp: O(n) thời gian, O(n) bộ nhớ.
# ============================================================

from typing import List


def find_max_length(nums: List[int]) -> int:
    """
    Tìm độ dài lớn nhất của subarray có số lượng 0 và 1 bằng nhau.

    Args:
        nums: Mảng chỉ gồm 0 và 1.

    Returns:
        Độ dài lớn nhất của subarray thỏa điều kiện.
    """

    # Map: prefix_sum -> index đầu tiên xuất hiện
    first_index = {0: -1}  # prefix_sum = 0 tại index -1

    prefix_sum = 0
    max_length = 0

    for i, num in enumerate(nums):

        # Chuyển đổi:
        # 0 -> -1, 1 -> +1
        if num == 0:
            prefix_sum -= 1
        else:
            prefix_sum += 1

        # Nếu prefix_sum đã xuất hiện trước đó
        if prefix_sum in first_index:
            length = i - first_index[prefix_sum]
            max_length = max(max_length, length)
        else:
            # Lưu index đầu tiên
            first_index[prefix_sum] = i

    return max_length


# ============================================================
# Test cases
# ============================================================

def run_tests():
    test_cases = [
        {
            "nums": [0, 1],
            "expected": 2,
            "desc": "Cả mảng có 1 số 0 và 1 số 1"
        },
        {
            "nums": [0, 1, 0],
            "expected": 2,
            "desc": "[0,1] hoặc [1,0]"
        },
        {
            "nums": [0, 0, 1, 0, 0, 0, 1, 1],
            "expected": 6,
            "desc": "Subarray dài nhất có số lượng 0 và 1 bằng nhau"
        },
        {
            "nums": [1, 1, 1, 1],
            "expected": 0,
            "desc": "Không có 0 → không có subarray hợp lệ"
        },
        {
            "nums": [0, 0, 0, 0],
            "expected": 0,
            "desc": "Không có 1 → không có subarray hợp lệ"
        },
        {
            "nums": [0, 1, 1, 0, 1, 1, 1, 0],
            "expected": 4,
            "desc": "Nhiều đoạn hợp lệ, lấy đoạn dài nhất"
        },
        {
            "nums": [0, 1, 0, 1, 0, 1],
            "expected": 6,
            "desc": "Cả mảng cân bằng hoàn toàn"
        },
    ]

    print("=" * 58)
    print("  Chạy test: Contiguous Array")
    print("=" * 58)

    passed = 0
    for i, tc in enumerate(test_cases):
        result = find_max_length(tc["nums"][:])
        ok = result == tc["expected"]
        if ok:
            passed += 1

        status = "PASS" if ok else "FAIL"
        print(f"\nTest {i + 1}: {tc['desc']}")
        print(f"  nums     : {tc['nums']}")
        print(f"  Expected : {tc['expected']}")
        print(f"  Got      : {result}")
        print(f"  Result   : [{status}]")

    print("\n" + "=" * 58)
    print(f"  Tổng kết: {passed}/{len(test_cases)} test passed")
    print("=" * 58)


if __name__ == "__main__":
    run_tests()
```
