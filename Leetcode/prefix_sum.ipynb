1. RUNNING SUM OF 1D ARRAY

```python
# ============================================================
# LeetCode 1480 - Running Sum of 1D Array
# Bài toán: Cho mảng nums, trả về mảng running sum.
# running_sum[i] = nums[0] + nums[1] + ... + nums[i]
# ============================================================

from typing import List


def running_sum(nums: List[int]) -> List[int]:
    """
    Tính running sum (tổng tích lũy) của mảng 1 chiều.

    Args:
        nums: Mảng số nguyên đầu vào.

    Returns:
        Mảng running sum cùng độ dài với nums.
    """

    # Duyệt qua từng vị trí từ index 1 đến cuối mảng
    # (index 0 giữ nguyên vì chưa có phần tử nào trước nó)
    for i in range(1, len(nums)):

        # Cộng dồn: phần tử hiện tại += phần tử liền trước
        # Sau bước này, nums[i] chứa tổng từ nums[0] đến nums[i]
        nums[i] += nums[i - 1]

    # Trả về mảng đã được biến đổi thành running sum
    return nums


# ============================================================
# Test cases
# ============================================================

def run_tests():
    test_cases = [
        {
            "input": [1, 2, 3, 4],
            "expected": [1, 3, 6, 10],
            "desc": "Mảng tăng dần thông thường"
        },
        {
            "input": [1, 1, 1, 1, 1],
            "expected": [1, 2, 3, 4, 5],
            "desc": "Tất cả phần tử bằng 1"
        },
        {
            "input": [3, 1, 2, 10, 1],
            "expected": [3, 4, 6, 16, 17],
            "desc": "Mảng hỗn hợp"
        },
        {
            "input": [5],
            "expected": [5],
            "desc": "Mảng chỉ có 1 phần tử"
        },
        {
            "input": [0, 0, 0],
            "expected": [0, 0, 0],
            "desc": "Toàn số 0"
        },
    ]

    print("=" * 50)
    print("  Chạy test: Running Sum of 1D Array")
    print("=" * 50)

    passed = 0
    for i, tc in enumerate(test_cases):
        # Tạo bản sao để không làm thay đổi dữ liệu gốc khi so sánh
        input_copy = tc["input"][:]
        result = running_sum(input_copy)
        ok = result == tc["expected"]

        status = "PASS" if ok else "FAIL"
        if ok:
            passed += 1

        print(f"\nTest {i + 1}: {tc['desc']}")
        print(f"  Input    : {tc['input']}")
        print(f"  Expected : {tc['expected']}")
        print(f"  Got      : {result}")
        print(f"  Result   : [{status}]")

    print("\n" + "=" * 50)
    print(f"  Tổng kết: {passed}/{len(test_cases)} test passed")
    print("=" * 50)


if __name__ == "__main__":
    run_tests()
```

2. FIND PIVOT INDEX

```python
# ============================================================
# LeetCode 724 - Find Pivot Index
# Bài toán: Tìm pivot index i sao cho:
#   sum(nums[0..i-1]) == sum(nums[i+1..n-1])
# Nếu không tìm được, trả về -1.
# ============================================================

from typing import List


def pivot_index(nums: List[int]) -> int:
    """
    Tìm chỉ số pivot của mảng.

    Ý tưởng:
        - Tính tổng toàn bộ mảng trước (total).
        - Duyệt từng index i, duy trì biến left_sum tích lũy.
        - Tổng bên phải = total - left_sum - nums[i]
        - Nếu left_sum == right_sum thì i là pivot.

    Độ phức tạp: O(n) thời gian, O(1) bộ nhớ.

    Args:
        nums: Mảng số nguyên đầu vào.

    Returns:
        Index pivot đầu tiên tìm được, hoặc -1 nếu không có.
    """

    # Tính tổng toàn bộ mảng một lần
    total = sum(nums)

    # Biến lưu tổng các phần tử bên trái index hiện tại
    # Ban đầu chưa có phần tử nào bên trái → left_sum = 0
    left_sum = 0

    for i in range(len(nums)):

        # Tổng bên phải = tổng toàn bộ - phần tử hiện tại - tổng bên trái
        right_sum = total - nums[i] - left_sum

        # Kiểm tra điều kiện pivot
        if left_sum == right_sum:
            return i  # Trả về ngay index đầu tiên tìm được

        # Cộng nums[i] vào left_sum để chuẩn bị cho vòng lặp tiếp theo
        left_sum += nums[i]

    # Duyệt hết mảng mà không tìm được pivot → trả về -1
    return -1


# ============================================================
# Test cases
# ============================================================

def run_tests():
    test_cases = [
        {
            "input": [1, 7, 3, 6, 5, 6],
            "expected": 3,
            "desc": "Pivot ở giữa mảng (index 3)"
        },
        {
            "input": [1, 2, 3],
            "expected": -1,
            "desc": "Không có pivot"
        },
        {
            "input": [2, 1, -1],
            "expected": 0,
            "desc": "Pivot ở đầu mảng (index 0, bên trái rỗng = 0)"
        },
        {
            "input": [-1, -1, -1, 0, 1, 1],
            "expected": 0,
            "desc": "Mảng có số âm, pivot ở index 0"
        },
        {
            "input": [0],
            "expected": 0,
            "desc": "Mảng 1 phần tử — luôn là pivot"
        },
        {
            "input": [0, 0, 0, 0],
            "expected": 0,
            "desc": "Toàn số 0 — pivot đầu tiên là index 0"
        },
        {
            "input": [-1, -1, 0, 1, 1, 0],
            "expected": 5,
            "desc": "Pivot ở cuối mảng (bên phải rỗng = 0)"
        },
    ]

    print("=" * 55)
    print("  Chạy test: Find Pivot Index")
    print("=" * 55)

    passed = 0
    for i, tc in enumerate(test_cases):
        result = pivot_index(tc["input"][:])  # bản sao để giữ input gốc
        ok = result == tc["expected"]
        if ok:
            passed += 1

        status = "PASS" if ok else "FAIL"
        print(f"\nTest {i + 1}: {tc['desc']}")
        print(f"  Input    : {tc['input']}")
        print(f"  Expected : {tc['expected']}")
        print(f"  Got      : {result}")
        print(f"  Result   : [{status}]")

    print("\n" + "=" * 55)
    print(f"  Tổng kết: {passed}/{len(test_cases)} test passed")
    print("=" * 55)


if __name__ == "__main__":
    run_tests()
```

3. CONTIGUOUS ARRAY

```python
# ============================================================
# LeetCode 525 - Contiguous Array
# Bài toán: Tìm độ dài lớn nhất của subarray có số lượng 0 và 1 bằng nhau.
#
# Ý tưởng: Biến bài toán về prefix sum.
#
# Ta coi:
#   0 -> -1
#   1 -> +1
#
# Khi đó:
# Nếu prefix_sum[j] == prefix_sum[i]
# → tổng từ i+1 đến j = 0
# → số lượng 0 và 1 bằng nhau
#
# Ta lưu vị trí đầu tiên xuất hiện của mỗi prefix_sum.
# Khi gặp lại cùng giá trị → cập nhật độ dài lớn nhất.
#
# Độ phức tạp: O(n) thời gian, O(n) bộ nhớ.
# ============================================================

from typing import List


def find_max_length(nums: List[int]) -> int:
    """
    Tìm độ dài lớn nhất của subarray có số lượng 0 và 1 bằng nhau.

    Args:
        nums: Mảng chỉ gồm 0 và 1.

    Returns:
        Độ dài lớn nhất của subarray thỏa điều kiện.
    """

    # Map: prefix_sum -> index đầu tiên xuất hiện
    first_index = {0: -1}  # prefix_sum = 0 tại index -1

    prefix_sum = 0
    max_length = 0

    for i, num in enumerate(nums):

        # Chuyển đổi:
        # 0 -> -1, 1 -> +1
        if num == 0:
            prefix_sum -= 1
        else:
            prefix_sum += 1

        # Nếu prefix_sum đã xuất hiện trước đó
        if prefix_sum in first_index:
            length = i - first_index[prefix_sum]
            max_length = max(max_length, length)
        else:
            # Lưu index đầu tiên
            first_index[prefix_sum] = i

    return max_length


# ============================================================
# Test cases
# ============================================================

def run_tests():
    test_cases = [
        {
            "nums": [0, 1],
            "expected": 2,
            "desc": "Cả mảng có 1 số 0 và 1 số 1"
        },
        {
            "nums": [0, 1, 0],
            "expected": 2,
            "desc": "[0,1] hoặc [1,0]"
        },
        {
            "nums": [0, 0, 1, 0, 0, 0, 1, 1],
            "expected": 6,
            "desc": "Subarray dài nhất có số lượng 0 và 1 bằng nhau"
        },
        {
            "nums": [1, 1, 1, 1],
            "expected": 0,
            "desc": "Không có 0 → không có subarray hợp lệ"
        },
        {
            "nums": [0, 0, 0, 0],
            "expected": 0,
            "desc": "Không có 1 → không có subarray hợp lệ"
        },
        {
            "nums": [0, 1, 1, 0, 1, 1, 1, 0],
            "expected": 4,
            "desc": "Nhiều đoạn hợp lệ, lấy đoạn dài nhất"
        },
        {
            "nums": [0, 1, 0, 1, 0, 1],
            "expected": 6,
            "desc": "Cả mảng cân bằng hoàn toàn"
        },
    ]

    print("=" * 58)
    print("  Chạy test: Contiguous Array")
    print("=" * 58)

    passed = 0
    for i, tc in enumerate(test_cases):
        result = find_max_length(tc["nums"][:])
        ok = result == tc["expected"]
        if ok:
            passed += 1

        status = "PASS" if ok else "FAIL"
        print(f"\nTest {i + 1}: {tc['desc']}")
        print(f"  nums     : {tc['nums']}")
        print(f"  Expected : {tc['expected']}")
        print(f"  Got      : {result}")
        print(f"  Result   : [{status}]")

    print("\n" + "=" * 58)
    print(f"  Tổng kết: {passed}/{len(test_cases)} test passed")
    print("=" * 58)


if __name__ == "__main__":
    run_tests()
```
